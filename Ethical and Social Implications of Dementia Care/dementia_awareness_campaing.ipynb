```python
import pandas as pd
import matplotlib.pyplot as plt


csv_file = 'dementia_awareness_campaing.csv' 
data = pd.read_csv(csv_file)
```

```python
awareness_campaigns = data[data['Existence_functioning_dementia_awareness_campaign_2017'] == 'Yes']
prevention_campaigns = data[data['Existence_functioning_risk_reduction_campaign_2017'] == 'Yes']

countries_awareness_campaign = len(awareness_campaigns)
countries_prevention_campaign = len(prevention_campaigns)

campaign_types = ['Awareness Campaigns', 'Prevention Campaigns']

campaign_counts = [countries_awareness_campaign, countries_prevention_campaign]

plt.figure(figsize=(8, 6))

plt.bar(campaign_types, campaign_counts, color=['blue', 'green'])
plt.xlabel('Campaign Types')
plt.ylabel('Number of Countries')
plt.title('Dementia Awareness and Prevention Campaigns Worldwide')
plt.ylim(0, max(campaign_counts) + 5)

for i, count in enumerate(campaign_counts):
    plt.text(i, count + 0.5, str(count), ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.show()
```

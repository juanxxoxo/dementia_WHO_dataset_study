```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('lack_physical_activity.csv')  # Replace with the actual file path

data[' Male'] = data[' Male'].apply(lambda x: float(x.split()[0]))
data[' Female'] = data[' Female'].apply(lambda x: float(x.split()[0]))

average_males = data.groupby('WHO_region')[' Male'].mean()
average_females = data.groupby('WHO_region')[' Female'].mean()


plt.figure(figsize=(10, 6))
average_males.plot(kind='bar', width=0.4, position=0, label='Male')
average_females.plot(kind='bar', width=0.4, position=1, color='orange', label='Female')

plt.title('Average Physical Exercise Between Men and Women by Region')
plt.xlabel('WHO Region')
plt.ylabel('Average Physical Exercise')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

plt.show()

```

```python
import pandas as pd
import re

data = pd.read_csv('lack_physical_activity.csv')  
def extract_numeric_value(text):
    match = re.search(r'([\d.]+)', text)
    if match:
        return float(match.group(1))
    return None

data[' Male'] = data[' Male'].apply(extract_numeric_value)
data[' Female'] = data[' Female'].apply(extract_numeric_value)

average_male_activity = data[data['Year'] == 2016][' Male'].mean()
average_female_activity = data[data['Year'] == 2016][' Female'].mean()

print("Average Male Activity in 2016:", round(average_male_activity, 2))
print("Average Female Activity in 2016:", round(average_female_activity, 2))

if average_female_activity < average_male_activity:
    print("Women performed less physical activity in 2016")
elif average_female_activity > average_male_activity:
    print("Men performed less physical activity in 2016")
else:
    print("Both genders performed the same amount of physical activity in 2016")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import re

data = pd.read_csv('lack_physical_activity.csv')  
def extract_numeric_value(text):
    match = re.search(r'([\d.]+)', text)
    if match:
        return float(match.group(1))
    return None

data[' Male'] = data[' Male'].apply(extract_numeric_value)
data[' Female'] = data[' Female'].apply(extract_numeric_value)

average_male_activity = data[data['Year'] == 2016][' Male'].mean()
average_female_activity = data[data['Year'] == 2016][' Female'].mean()

plt.figure(figsize=(6, 4))
plt.bar(['Male', 'Female'], [average_male_activity, average_female_activity], color=['blue', 'orange'])
plt.title('Average Physical Exercise Between Men and Women in 2016')
plt.ylabel('Average Physical Exercise')
plt.tight_layout()

if average_female_activity < average_male_activity:
    print("Women performed less physical activity in 2016")
elif average_female_activity > average_male_activity:
    print("Men performed less physical activity in 2016")
else:
    print("Both genders performed the same amount of physical activity in 2016")

plt.show()
```

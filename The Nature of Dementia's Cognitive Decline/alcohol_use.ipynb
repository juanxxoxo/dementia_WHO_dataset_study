```python
import pandas as pd
import matplotlib.pyplot as plt

csv_file = 'alcohol_use.csv'  
data = pd.read_csv(csv_file)
```

```python
# Calculate the average consumption for the last 5 years
data['average_last_5_years'] = (
    data[['alcohol_per_capita_consumption_2009', 'alcohol_per_capita_consumption_2008',
          'alcohol_per_capita_consumption_2007', 'alcohol_per_capita_consumption_2006',
          'alcohol_per_capita_consumption_2005']].sum(axis=1) / 5
)

# Calculate the average consumption for the first 5 years
data['average_first_5_years'] = (
    data[['alcohol_per_capita_consumption_2004', 'alcohol_per_capita_consumption_2003',
          'alcohol_per_capita_consumption_2002', 'alcohol_per_capita_consumption_2001',
          'alcohol_per_capita_consumption_2000']].sum(axis=1) / 5
)
```

```python
average_diff = data['average_last_5_years'].mean() - data['average_first_5_years'].mean()

plt.figure(figsize=(8, 6))

plt.bar(['Average Difference'], [average_diff], color='blue')
plt.xlabel('Metric')
plt.ylabel('Average Difference')
plt.title('Average Alcohol Consumption Difference (Last 5 Years vs First 5 Years)')
plt.ylim(0, average_diff + 0.2)  

plt.text(0, average_diff + 0.05, f'{average_diff:.2f}', ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.show()
```
